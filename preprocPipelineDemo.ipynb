```python
import os
import numpy as np
import pandas as pd
import scipy as sci
import tifffile as tif
import random

import matplotlib.pyplot as plt
```

```python
fileName ='DataSet_CdTe/interstitial_50.tif'
```

```python
image = tif.imread(fileName)[:,:,0]
plt.imshow(image)
#plt.imsave('figures/preProc1.png', image)
```

```python
#Makes all images scaled the same
image = image / np.max(image)
plt.imshow(image)
#plt.imsave('figures/preProc2.png', image)
```

```python
image =  sci.ndimage.gaussian_filter(image, 1)
plt.imshow(image)
#plt.imsave('figures/preProc3.png', image)
```

kernal = np.array([
    [1, 0, -1],
    [1, 0, -1],
    [1, 0, -1]
])
image = sci.signal.convolve2d(image, kernal)
plt.imshow(image)

```python
# division safe fft
image = np.log(1 + np.abs(np.fft.fftshift(np.fft.fft2(image))))
plt.imshow(image)
#plt.imsave('figures/preProc4.png', image)
```

```python
image = (image > 0.1) * image
plt.imshow(image)
#plt.imsave('figures/preProc5.png', image)
```

```python
image = (image < 7) * image
plt.imshow(image)
#plt.imsave('figures/preProc5.png', image)
```
